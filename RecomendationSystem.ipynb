```python
user_book_df = book.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating').reset_index(drop=True)
```

```python
user_book_df
```

```python

user_book_df.index = book['User.ID'].unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python

# Calculating Cosine Similarity between Users on array data
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python

user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index = book['User.ID'].unique()
user_sim_df.columns = book['User.ID'].unique()
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)
```
